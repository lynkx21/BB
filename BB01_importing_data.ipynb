```python
from scipy import stats, linalg

def partial_corr(C):
    """
    Returns the sample linear partial correlation coefficients between pairs of variables in C, controlling 
    for the remaining variables in C.
    Parameters
    ----------
    C : array-like, shape (n, p)
        Array with the different variables. Each column of C is taken as a variable
    Returns
    -------
    P : array-like, shape (p, p)
        P[i, j] contains the partial correlation of C[:, i] and C[:, j] controlling
        for the remaining variables in C.
    """

    C = np.asarray(C)
    p = C.shape[1]
    P_corr = np.zeros((p, p), dtype=np.float)
    for i in range(p):
        P_corr[i, i] = 1
        for j in range(i+1, p):
            idx = np.ones(p, dtype=np.bool)
            idx[i] = False
            idx[j] = False
            beta_i = linalg.lstsq(C[:, idx], C[:, j])[0]
            beta_j = linalg.lstsq(C[:, idx], C[:, i])[0]

            res_j = C[:, j] - C[:, idx].dot( beta_i)
            res_i = C[:, i] - C[:, idx].dot(beta_j)

            corr = stats.pearsonr(res_i, res_j)[0]
            P_corr[i, j] = corr
            P_corr[j, i] = corr

    return P_corr

partial_corr_array = df_kpis.drop(['Date', 'year', 'month', 'weekday'], axis=1)

# Calculate the partial correlation coefficients
a = partial_corr(partial_corr_array[['K3', 'K5', 'K9']])
```

```python
a = pd.DataFrame(a,
                 columns=df_kpis.drop(['Date', 'year', 'month', 'weekday'], axis=1)[['K3', 'K5', 'K9']].columns,
                 index=df_kpis.drop(['Date', 'year', 'month', 'weekday'], axis=1)[['K3', 'K5', 'K9']].columns)
```
